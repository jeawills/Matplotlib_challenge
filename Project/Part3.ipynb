```python
# Import the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bring each CSV into a separate data frame
data2013_df = "Resources/WWE-Data-2013.csv" 
data2013_df = pd.read_csv(data2013_df)
data2013_df.head()
```

```python
data2014_df = "Resources/WWE-Data-2014.csv" 
data2014_df = pd.read_csv(data2014_df)
data2014_df.head()
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
data_13_14_df = pd.merge(data2013_df, data2014_df, on="Wrestler", how="outer")
data_13_14_df
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws"
data_13_14_df = data_13_14_df.rename(
    columns={"Wins_x": "2013 Wins", "Losses_x": "2013 Losses", "Draws_x": "Draws 2013", "Wins_y": "2014 Wins", "Losses_y": "2014 Losses", "Draws_y": "2014 Draws" })
data_13_14_df.head()

# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"

```

```python
# Bring each CSV into a separate data frame
data2015_df = "Resources/WWE-Data-2015.csv" 
data2015_df = pd.read_csv(data2015_df)
data2015_df.head()
```

```python
# Merge our newly combined dataframe with the 2015 dataframe
data_13_14_15_df = pd.merge(data_13_14_df, data2015_df, on="Wrestler", how="outer")
data_13_14_15_df
```

```python
# Rename "wins", "losses", and "draws" to "2015 Wins", "2015 Losses", and "2015 Draws"
data_13_14_15_df = data_13_14_15_df.rename(
    columns={"Wins": "2015 Wins", "Losses": "2015 Losses", "Draws": "2015 Draws"})
data_13_14_15_df.head()
```

```python
# Rename "wins", "losses", and "draws" to "2015 Wins", "2015 Losses", and "2015 Draws"
data_13_14_15_df = data_13_14_15_df.rename(
    columns={"Wins": "2015 Wins", "Losses": "2015 Losses", "Draws": "Draws 2015"})
data_13_14_15_df.head()
```

```python
data2016_df = "Resources/WWE-Data-2016.csv" 
data2016_df = pd.read_csv(data2016_df)
data2016_df.head()
```

```python
# Merge our newly combined dataframe with the 2016 dataframe
data_13_14_15_16_df = pd.merge(data_13_14_15_df, data2016_df, on="Wrestler", how="outer")
data_13_14_15_16_df
```

```python
# Rename "wins", "losses", and "draws" to "2016 Wins", "2016 Losses", and "2016 Draws"
data_13_14_15_16_df = data_13_14_15_16_df.rename(
    columns={"Wins": "2016 Wins", "Losses": "2016 Losses", "Draws": "Draws 2016"})
data_13_14_15_16_df.head()
```

```python
#Part 2 - Time to Calculate!
#When your tables have been merged together into one data frame, calculate the total number of wins, losses, and draws a wrestler has had over the course of their career. Also create a new column that will hold the total matches a wrestler has been in over the course of their career.

#You will need to convert all NaN values to a number so that you can perform these calculations
#We are only interested in those wrestlers who have been with the WWE from 2013 to 2016. You will need to come up with some way of filtering out rows that do not meet these conditions.

#Also set the 'Wrestler' column as your key for easier referencing later on.
```

```python
# Replace all NaN values with 0 

# Create a new column called "Total Wins" and add up each wrestler's wins per year to fill in the values

# Create a new column called "Total Losses" and add up each wrestler's losses per year to fill in the values

# Create a new column called "Total Draws" and add up each wrestler's draws per year to fill in the values

# Create a new column called "Total Matches" and add up the total wins, losses, and draws for each wrestler to fill in the values
```

```python
# Replace all NaN values with 0 
combined_data_df = data_13_14_15_16_df 
combined_data_df = combined_data_df.fillna(0)
combined_data_df 
```

```python
# Create a new column called "Total Wins" and add up each wrestler's wins per year to fill in the values
combined_data_df["Total Wins"] = combined_data_df["2013 Wins"] + combined_data_df["2014 Wins"] + combined_data_df["2015 Wins"] + combined_data_df["2016 Wins"]
combined_data_df
```

```python
# Create a new column called "Total Losses" and add up each wrestler's losses per year to fill in the values
combined_data_df["Total Losses"] = combined_data_df["2013 Losses"] + combined_data_df["2014 Losses"] + combined_data_df["2015 Losses"] + combined_data_df["2016 Losses"]
combined_data_df
```

```python
# Create a new column called "Total Draws" and add up each wrestler's draws per year to fill in the values
combined_data_df["Total Draws"] = combined_data_df["Draws 2013"] + combined_data_df["2014 Draws"] + combined_data_df["2015 Draws"] + combined_data_df["Draws 2016"]
combined_data_df
```

```python
# Create a new column called "Total Matches" and add up the total wins, losses, and draws for each wrestler to fill in the values
combined_data_df["Total Matches"] = combined_data_df["Total Wins"] + combined_data_df["Total Losses"] + combined_data_df["Total Draws"] 
combined_data_df
```

```python
# Create a new dataframe for those wrestlers who have wrestled at least 100 matches,
wrestled_over_hundred = combined_data_df.loc[(combined_data_df["Total Matches"] >= 100) &
                                                 (combined_data_df["2013 Wins"] > 0) &
                                                 (combined_data_df["2016 Wins"] > 0)]

# Set the index of this new dataframe to be the wrestlers names
wrestled_over_hundred = wrestled_over_hundred.set_index("Wrestler")

wrestled_over_hundred.head()
```

Part 3 - Charting Careers
Store an individual wrestler's wins over time in a variable
Store that same wrestler's losses over time in a variable as well
Create a line chart that will plot this wrestler's wins and losses from 2013 to 2016

```python
# Collect the user's input to search through our data frame
wrestler_name = input("Which wrestler's career would you like to look at? ")
```

```python
# Create a series that looks for a wrestler by name and then traces their wins from 2013 to 2016
wins = wrestled_over_hundred.loc[wrestler_name, ['2013 Wins', '2014 Wins', '2015 Wins', '2016 Wins']]
wins
```

```python
# Create a series that looks for a wrestler by name and then traces their losses from 2013 to 2016
losses = wrestled_over_hundred.loc[wrestler_name, ['2013 Losses', '2014 Losses', '2015 Losses', '2016 Losses']]
losses
```

```python
# Create a list of the years that we will use as our x axis
years = [2013, 2014, 2015, 2016]

# Plot our line that will be used to track a wrestler's wins over the years
plt.plot(years, wins, color="green", label="Wins")

# Plot our line that will be used to track a wrestler's losses over the years
plt.plot(years, losses, color="red", label="Losses")

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

plt.title(wrestler_name + " 's Career Snapshot")
plt.xlabel("Years")
plt.ylabel("Wins/Losses")

# Print our chart to the screen
plt.show()
```
